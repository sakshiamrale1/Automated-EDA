## Analysis of Data:

### EDA on Numerical DATA:

```python
num_df = df.select_dtypes(include=np.number)
```

```python
Total_List = []
for i in range(len(num_df.columns)):
    Total_List.append(num_df.iloc[:,i].sum())
    

x = num_df.columns
y = Total_List

print(x)
print()
plt.bar(x,y)
plt.show()
```

## EDA on Non-Numerical DATA:

```python
non_num_df = df.select_dtypes(exclude=np.number)
non_num_df.columns
```

Unique Values of Each Categorical Column:

```python
for i in non_num_df.columns:
    print("Unique Values for column ",i," are:")
    print(df[i].unique())
    print()
    print()
```

Value Count of Each Categorical Column:

```python
for i in non_num_df.columns:
    print("Value Count for column ",i," is:")
    Val_Counts = df[i].value_counts()
    print(Val_Counts)
    print()
    print()
    
```

Percentage wise Analysis of Each categorical Column:

```python

for i in non_num_df.columns:
    if len(non_num_df[i].value_counts().keys()) < 20:
        y = non_num_df[i].value_counts().values
        mylabels = non_num_df[i].value_counts().keys()
        
        plt.pie(y, labels = mylabels,autopct='%.2f')
        plt.title(i)
        plt.show() 
    
```

Plotting Barplots for each categorical column using value Counts:

```python
for i in non_num_df.columns:
    if len(non_num_df[i].value_counts().keys()) < 20:
        Val_Counts = df[i].value_counts()
        sns.barplot(Val_Counts.index , Val_Counts.values)
        plt.xticks(rotation=90)
        plt.xlabel(i)
        plt.ylabel('Count')
        plt.show()
        
```

## Extra EDA:

```python
df['Year'] = pd.DatetimeIndex(df['Order Date']).year
df.head(2)
```

```python
df['Month'] = pd.DatetimeIndex(df['Order Date']).month
df.head(2)
```

```python
year1 = df['Year'].value_counts()
year1
```

```python
plt.figure(figsize = (16,5))
sns.barplot(year1.index, year1.values)
plt.show()
```

Monthly Sale

```python
m2 = df.groupby("Month").count()
m2
```

Yearly Sale
